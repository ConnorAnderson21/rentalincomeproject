```python



class Rental_property():
    
#     create attributes for each box in his model from video
#     create a function that takes in input which is combined to determine value of attributes
#     


    def __init__(self):
        self.income = 0.0
        self.expenses = 0.0
        self.cashflow = 0.0
        self.totinv = 0.0
        self.roi = 0.0
        
    def calc_income(self):
        rent = input("enter monthly rent")
        storage = input("enter monthly storage fees")
        laundry = input("enter monthly laundry income")
        misc = input("enter monthly misc income")
        self.income = float(rent) + float(storage) + float(laundry) + float(misc)
        print(f"Your monthly income is: ${self.income}0")
        
    def calc_expenses(self):
        tax = input("enter monthly tax cost")
        insurance = input("enter monthly insurance cost")
        utilities = input("enter monthly utilities cost")
        hoa = input("enter monthly hoa cost")
        lawn = input("enter monthly lawncare cost")
        vacancy = input("enter monthly vacancy cost")
        repairs = input("enter monthly repairs cost")
        capex = input("enter monthly capex cost")
        propmanage = input("enter monthly property management cost")
        mortgage = input("enter monthly mortgage cost")
        self.expenses = float(tax) + float(insurance) + float(utilities) + float(hoa) + float(lawn) + float(vacancy) + float(repairs) + float(capex) + float(propmanage) + float(mortgage)
        print(f"Your monthly expenses are: ${self.expenses}0")
        
# Could this perhaps be made more concise using a dictionary?....

    def calc_cashflow(self):
        self.cashflow = float(self.income) - float(self.expenses)
        print(f"Your monthly cash flow is: ${self.cashflow}0")
    
    def calc_totinv(self):
        down_payment = input("enter down payment")
        closing = input("enter closing costs")
        rehab = input("enter rehab costs")
        misc = input("enter misc costs")
        self.totinv = float(down_payment) + float(closing) + float(rehab) + float(misc)
        print(f"Your total investment is: ${self.totinv}0")
        
    def calc_roi(self):
        self.roi = self.cashflow * 12 / self.totinv * 100
        print(f"Your return on investment is: {self.roi}%")
        
def run(self):
        while True:
            response = input('Would you like to calculate the return on investment for a property? yes or no')
            if response == 'yes':
                self.calc_income()
                self.calc_expenses()
                self.calc_cashflow()
                self.calc_totinv()
                self.calc_roi()
                break
            elif response == 'no':
                print("Oh come on! I'm just trying to help you get rich!")
            else:
                print("Please answer the question.")
            
    
    
    
    
    
    
#     while 
        
        
#         need a menu 
        

new_house = Rental_property()
    


run(new_house)
        



```
